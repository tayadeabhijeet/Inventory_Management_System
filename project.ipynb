# importing file and some operations

```python
import json

fd = open("inventory.json",'r')
r = fd.read()
fd.close()

inventory = json.loads(r)
```

```python
inventory
```

# Buy

```python
ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))

print("Product: ", inventory[ui_prod]['name'])
print("Price: ", inventory[ui_prod]['price'])
print("Brand:", inventory[ui_prod]['brand'])
print("Expiry Date")
print("Billing Amount: ", inventory[ui_prod]['price'] * ui_quant)

inventory[ui_prod]['quantity'] = inventory[ui_prod]['quantity'] - ui_quant
```

```python
js = json.dumps(inventory)

fd = open("inventory.json",'w')
fd.write(js)
fd.close()
```

# Add New Item into Inventory

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
price = int(input("Enter price:"))
brand = str(input("Enter brand:"))
quantity = int(input("Enter quantity:"))

inventory[prod_id] = {'name': name, 'price': price, 'quantity': quantity, 'brand': brand}

js = json.dumps(inventory)

fd = open("inventory.json",'w')
fd.write(js)
fd.close()
```
